# Question 1 – Sentence and Word Tokenization

## Objective
To perform sentence-level and word-level tokenization on the corpus and compute fundamental corpus statistics including:
- Total number of tokens (N)
- Total vocabulary size / types (V)
- Type-Token Ratio (TTR)
- Top 10 most frequent tokens

## Methodology

### Step 1: Corpus Construction
All 2000 reviews are combined into one large string to treat the dataset as a single corpus for global analysis.

### Step 2: Tokenization
- `sent_tokenize()` is used to split the corpus into sentences.
- `word_tokenize()` is used to split the corpus into word-level tokens.

### Step 3: Statistical Computations
The following metrics are computed:

- **Total Tokens (N):** Total number of word tokens in the corpus.
- **Total Types (V):** Number of unique tokens.
- **Type-Token Ratio (TTR):** A measure of lexical diversity.

Type-Token Ratio is calculated as:

TTR = V / N

A TTR value closer to **1** indicates high lexical diversity, while a value closer to **0** indicates high repetition.

### Step 4: Frequency Analysis
Token frequencies are calculated using `Counter`, and the top 10 most frequent tokens are displayed.

---

## Output Interpretation

- **Number of Sentences:** 9252  
- **Total Tokens (N):** 177,078  
- **Vocabulary Size (V):** 11,543  
- **Type-Token Ratio (TTR):** 0.0652  

### Observations
1. The TTR (0.0652) is closer to **0**, indicating repetition is common in the corpus.
2. The most frequent tokens are punctuation marks and stopwords (e.g., ".", ",", "the", "I"), which is expected in raw text.
3. Punctuation appears among the most frequent tokens because no preprocessing has been applied yet.

This completes the baseline corpus analysis before applying stopword removal, stemming, or lemmatization.

```python
from nltk.tokenize import sent_tokenize, word_tokenize
from collections import Counter

nltk.download('punkt', quiet=True)

# combine all reviews into one large string
all_text = " ".join(texts.tolist())

# sentence & word tokenization
sentences = sent_tokenize(all_text)
word_tokens = word_tokenize(all_text)

# compute statistics
N = len(word_tokens)
V = len(set(word_tokens))
TTR = V / N if N > 0 else 0

freq = Counter(word_tokens)
top10 = freq.most_common(10)

print(f"Number of sentences: {len(sentences)}")
print(f"Total tokens (N): {N}")
print(f"Total Vocabulary Size (V): {V}")
print(f"Type‑Token Ratio (TTR): {TTR:.4f}")
print("\nTop 10 most frequent tokens:")
for token, count in top10:
    print(f"{token}: {count}")
```

# Question 3 – Porter Stemming and Vocabulary Reduction

## Objective
To apply **Porter Stemming** to the corpus tokens and analyze its impact on vocabulary size.

## Methodology

### Step 1: Apply Porter Stemmer
Each token in the filtered corpus is processed using `PorterStemmer()` to reduce words to their root form (stem).

Example:
- playing → play  
- played → play  
- studies → studi  

Stemming removes suffixes but does not guarantee linguistically valid words.

### Step 2: Compute Vocabulary After Stemming
The new vocabulary size is calculated by counting unique stemmed tokens.

### Step 3: Calculate Percentage Reduction
Vocabulary reduction is computed using:

Reduction (%) = ((V_before − V_after) / V_before) × 100

Where:
- V_before = Vocabulary after stopword removal
- V_after = Vocabulary after stemming

---

## Output Interpretation

- **Vocabulary after stemming:** 7,315  
- **Percentage reduction:** 34.85%

### Observations
1. Stemming reduced vocabulary size by approximately **35%**.
2. This indicates that many words share common morphological roots.
3. Stemming significantly compresses the vocabulary space.
4. However, stems may not always be real words (e.g., *studi* instead of *study*).

---

## Conclusion
- Stemming substantially reduces vocabulary size.
- It improves normalization but may distort word forms.
- Vocabulary reduction is significant, confirming that morphological variations contribute heavily to vocabulary diversity.

```python
from nltk.stem import PorterStemmer

stemmer = PorterStemmer()

stemmed_tokens = [stemmer.stem(w) for w in filtered_tokens]

V_stem = len(set(stemmed_tokens))

reduction_percent = ((V2 - V_stem) / V2) * 100

print("Vocabulary after stemming:", V_stem)
print("Percentage reduction:", round(reduction_percent, 2), "%")
```

# Question 7 – Complete Text Preprocessing Pipeline (Review-Level Analysis)

## Objective
To demonstrate a **complete NLP preprocessing pipeline** on three selected reviews, showing the full transformation of text at each stage.

Each review is processed step-by-step to observe how text length, structure, and vocabulary change throughout the pipeline.

---

## Preprocessing Steps

For each selected review, the following transformations are applied **in sequence**, with full output displayed at every stage:

1. **Original Review**  
   Raw text as present in the dataset.

2. **HTML Removal**  
   Removes any HTML tags using regular expressions.

3. **Punctuation Normalization**  
   - `!!!` → `!`  
   - `???` → `?`  
   - `...` → `.`  
   This reduces punctuation noise while preserving meaning.

4. **Lowercasing**  
   Converts all text to lowercase to ensure case-insensitive processing.

5. **Tokenization**  
   Splits text into individual word tokens using `word_tokenize`.

6. **Stopword Removal (Negation Preserved)**  
   Removes common stopwords while explicitly keeping:
   - no  
   - not  
   - never  
   - n't  

7. **Lemmatization (pos='v')**  
   Converts tokens to their base verb forms using WordNet lemmatization.

---

## Why This Step Matters
- Shows **end-to-end text transformation**.
- Highlights how preprocessing reduces noise.
- Demonstrates semantic preservation despite heavy normalization.
- Fulfills all requirements of **Question 7 instructions**.

The full output for each stage is printed for **all three reviews**, without skipping any step.

```python
def preprocess_review(review):

    print("="*80)
    print("1️⃣ Original Review:\n")
    print(review)

    # 2. Remove HTML
    no_html = re.sub(r'<.*?>', '', review)
    print("\n2️⃣ After HTML Removal:\n")
    print(no_html)

    # 3. Normalize punctuation
    normalized = re.sub(r'!+', '!', no_html)
    normalized = re.sub(r'\?+', '?', normalized)
    normalized = re.sub(r'\.+', '.', normalized)
    print("\n3️⃣ After Punctuation Normalization:\n")
    print(normalized)

    # 4. Lowercasing
    lower = normalized.lower()
    print("\n4️⃣ After Lowercasing:\n")
    print(lower)

    # 5. Tokenization
    tokens = word_tokenize(lower)
    print("\n5️⃣ After Tokenization:\n")
    print(tokens)

    # 6. Stopword removal (negation preserved)
    filtered = [w for w in tokens if w not in stop_words]
    print("\n6️⃣ After Stopword Removal:\n")
    print(filtered)

    # 7. Lemmatization (verb)
    lemmatized = [lemmatizer.lemmatize(w, pos='v') for w in filtered]
    print("\n7️⃣ After Lemmatization (pos='v'):\n")
    print(lemmatized)
```
